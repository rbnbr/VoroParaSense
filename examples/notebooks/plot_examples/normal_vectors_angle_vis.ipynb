```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors
vector1 = np.array([2, 3])
vector2 = np.array([-1, 2])

# Normalize the vectors
vector1 = vector1 / np.linalg.norm(vector1)
vector2 = vector2 / np.linalg.norm(vector2)

# Calculate the angle between the vectors
angle = np.arccos(np.dot(vector1, vector2))
angle_degrees = np.degrees(angle)

# Create the plot
fig, ax = plt.subplots()
ax.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector 1')
ax.quiver(0, 0, vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='green', label='Vector 2')

# Plot orthogonal lines as dashed lines
orthogonal1 = np.array([vector1[1], -vector1[0]])
orthogonal2 = np.array([vector2[1], -vector2[0]])
ax.plot([-orthogonal1[0], orthogonal1[0]], [-orthogonal1[1], orthogonal1[1]], 'r--', label='Orthogonal to Vector 1')
ax.plot([-orthogonal2[0], orthogonal2[0]], [-orthogonal2[1], orthogonal2[1]], 'm--', label='Orthogonal to Vector 2')

# Annotate the angle
# ax.text(0.5, 0.5, f'Angle: {angle_degrees:.2f}°', fontsize=12, ha='center')

# Set limits and labels
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()
ax.grid(True)
ax.set_aspect('equal', 'box')
fig.tight_layout()
fig.savefig("vector_vector_angle.pdf")

plt.show()
```
