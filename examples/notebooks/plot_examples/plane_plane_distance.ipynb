```python
import numpy as np
```

```python
def plane_angle_metric(v1: np.ndarray, v2: np.ndarray) -> float:
    """
    """
    v1l = np.linalg.norm(v1)
    v2l = np.linalg.norm(v2)
    assert v1l > 0
    assert v2l > 0

    if v1l != 1.0:
        v1 = v1 / v1l

    if v2l != 1.0:
        v2 = v2 / v2l

    v = np.abs(v1 @ v2).item()
    if v < -1.0:
        v = -1.0
    elif v > 1.0:
        v = 1.0

    return np.arccos(v) / (np.pi * 0.5)
```

```python
def plane_angle_metric2(v1: np.ndarray, v2: np.ndarray) -> float:
    """
    """
    v1l = np.linalg.norm(v1)
    v2l = np.linalg.norm(v2)
    assert v1l > 0
    assert v2l > 0

    if v1l != 1.0:
        v1 = v1 / v1l

    if v2l != 1.0:
        v2 = v2 / v2l

    v = (v1 @ v2).item()

    return 1 - abs(v)
```

```python
plane_angle_metric(np.array([1, 0, 0]), np.array([1, 0, 1]))
```

```python
plane_angle_metric2(np.array([1, 0, 0]), np.array([0, 1, 0]))
```

```python
x = np.random.random((10, 3))
```

```python
for x_ in x:
    print(plane_angle_metric(x_, x[0]), plane_angle_metric2(x_, x[0]))
```
